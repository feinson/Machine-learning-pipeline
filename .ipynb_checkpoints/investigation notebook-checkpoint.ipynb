```python
import pandas as pd
import numpy as np

df = pd.read_csv('./data/clean_tabular_data.csv')

print(df.dtypes)
print(len(df.columns))
df.head()
```

```python
def load_airbnb(clean_df: pd.DataFrame, label):
    df = clean_df.copy()
    labels = np.array(df.pop(label))
    features = np.array(df.select_dtypes(['number']))
    return (features, labels)
```
